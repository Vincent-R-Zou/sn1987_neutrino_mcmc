```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter  # Import the smoothing filter


E_K, eta_k = np.loadtxt(r"F:/Neutrino_SI/Bin/eff_Kam_II.csv", delimiter=',', unpack=True)
E_I, eta_i = np.loadtxt(r"F:/Neutrino_SI/Bin/eff_IMB.csv", delimiter=',', unpack=True)
E_B, eta_b = np.loadtxt(r"F:/Neutrino_SI/Bin/eff_B.csv", delimiter=',', unpack=True)

fig, ax = plt.subplots(figsize=(8, 6))

v1 = ax.axvline(x=4, label='Kam threshold', ls='--', color='blue', alpha=0.2)
v2 = ax.axvline(x=12, label='Baksan threshold',ls='--', color='green', alpha=0.2)
v3 = ax.axvline(x=17, label='IMB threshold', ls='--', color='red', alpha=0.2)

ax.plot(E_K, eta_k, label = "Kam-II", color='blue')
ax.plot(E_I, eta_i, label = "IMB", color='red')
ax.plot(E_B, eta_b, label = "Baksan", color='green')

ax.set_xlim(0, 50)
ax.set_ylim(0, 1)
ax.set_xlabel(r"$E_\nu$ (MeV)")
ax.set_ylabel(r"Detection efficiency")
ax.set_title(r"Detection efficiency of Kam-II, IMB, and Baksan")

plt.legend()

```
