```python
from Pycode import neutrino
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt


m_phi_range = np.linspace(0, 1500, 1000)  # eV
E_k_range = np.linspace(5, 50, 1000)      # MeV

M, E = np.meshgrid(m_phi_range, E_k_range)

G = neutrino.g_function(M, E)

plt.figure(figsize=(6,5), dpi=300)
im = plt.pcolormesh(M, E, G, shading='auto', cmap='plasma',)
plt.colorbar(im, label=r'$m_\phi*g(m_\phi, E_k)$')
plt.xlabel(r'$m_\phi$ (eV)')
plt.ylabel(r'$E_k$ (MeV)')
plt.tight_layout()
```
